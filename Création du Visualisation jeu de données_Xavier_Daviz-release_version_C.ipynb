```python
# importer la fonction chi2_contingency  la correlation entre la variable Home Team et la variable Cible FTR
from scipy.stats import chi2_contingency
# réaliser le test du X2 sur la table de contingency table_HOME
resultat_test =chi2_contingency(table_HOME)
statistique =resultat_test [0]
p_value = resultat_test [1]
degre_liberte =resultat_test [2]

#afficher la statistique du test, la p__value , et le degre de liberté

print( "la statistique:",statistique,",p_value:",p_value, " ,le degre de liberté :", degre_liberte)
```
